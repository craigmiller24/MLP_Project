```python
# import necessary modules
import pandas as pd
import random
import torch

from baseline_models import train_multinomial_logistic_regression, train_decision_tree
from benchmark_models import train_lstm, train_random_forest, train_rf_then_lstm
from cross_validation import crossvalidation_training, metric_avgs
```

```python
# Set a fixed random seed for reproducibility
SEED = 50
random.seed(SEED)
torch.manual_seed(SEED)
```

```python
# Importing datasets

merged_jfk_df = pd.read_csv('data/complete/JFK_final_df.csv')
merged_mia_df = pd.read_csv('data/complete/MIA_final_df.csv')
merged_lax_df = pd.read_csv('data/complete/LAX_final_df.csv')
merged_ewr_df = pd.read_csv('data/complete/EWR_final_df.csv')
merged_ord_df = pd.read_csv('data/complete/ORD_final_df.csv')
```

```python
# Airport data
airports = ['JFK', 'EWR', 'LAX', 'MIA', 'ORD']
airport_data_dict = {
    'JFK': merged_jfk_df,
    'EWR': merged_ewr_df,
    'LAX': merged_lax_df,
    'MIA': merged_mia_df,
    'ORD': merged_ord_df
}
```

```python
# Training MLR model
mlr_metrics = crossvalidation_training(train_multinomial_logistic_regression, airports, airport_data_dict)
mlr_avg_metrics = metric_avgs(mlr_metrics)
```

```python
# Training DT model
dt_metrics = crossvalidation_training(train_decision_tree, airports, airport_data_dict)
dt_avg_metrics = metric_avgs(dt_metrics)
```

```python
# Training RF model
rf_metrics = crossvalidation_training(train_random_forest, airports, airport_data_dict)
```

```python
# Training LSTM model
lstm_metrics = crossvalidation_training(train_lstm, airports, airport_data_dict)
```

```python
# Training Ensemble model
ensemble_metrics = crossvalidation_training(train_rf_then_lstm, airports, airport_data_dict)
```

```python
print(f'MLR Metrics: {mlr_metrics}')
print(f'DT Metrics: {dt_metrics}')
print(f'RF Metrics: {rf_metrics}')
print(f'LSTM Metrics: {lstm_metrics}')
print(f'Ensemble Metrics: {ensemble_metrics}')
```
